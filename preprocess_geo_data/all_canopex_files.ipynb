```python
def correct_spaces(f_name):
    # f_name: (str) file name
    lines = []
    prev_line=''
    f = open(f_name, 'r')

    try:
        
        for line in f:
            line = line.strip()
            if (line[4] != "\t"):
                line = line[:4] + "\t" + line[4:]
            if (line[6] == '1'):
                if ( (int(prev_line[6]) == 9) | (int(prev_line[6:8]) == 10) | (int(prev_line[6:8]) == 11) ):
                    next
                elif (int(line[6:8]) == 12) & (int(prev_line[6:8]) == 12):
                    next
                elif (prev_line[6:10] == '1231'):
                    line = line[:6] + str(0) + line[6:]
                elif (int(prev_line[6:8]) == 1):
                    line = line[:6] + str(0) + line[6:]

            elif (line[6] != '1'):
                line = line[:6] + str(0) + line[6:]

            prev_line = line


            if (line[8] != "\t"):
                 line = line[:8] + "\t" + line[8:]
            #print(line)

            lines.append(line)

        f.close()
        
    except IndexError:
        f.close()
        pass
    
    return lines
```

```python
def preprocess(lines, f_name):
    year = []
    month = []
    day = []
    precip = []
    evap = []
    discharge = []
    max_temp = []
    min_temp = []

    for line in lines:
        year.append(int(line[0:4]))

        if (int(line[5:7]) == 11):
            month.append(11)
        else:
            try:
                month.append(int((line[5]+line[7])))
            except ValueError:
                month.append(1)

        day.append(int(line[9:11]))
        precip.append(float(line[15:21]))
        evap.append(float(line[23:31]))
        discharge.append(float(line[34:41]))
        max_temp.append(float(line[43:51]))
        min_temp.append(float(line[53:61]))

    year = np.array(year)
    month = np.array(month)
    day = np.array(day)
    precip = np.array(precip)
    evap = np.array(evap)
    discharge = np.array(discharge)
    max_temp = np.array(max_temp)
    min_temp = np.array(min_temp)

    dates = []
    for i in range(0, len(year)):
        date = pd.Timestamp(year=year[i], month=month[i], day=day[i])
        dates.append(date)
    dates = np.array(dates)

    discharge[np.where(discharge == 99.0)] = np.nan
    evap[np.where(evap == -99.0)] = np.nan

    d = {'date': dates, 'precip': precip, 'evap': evap, 'discharge': discharge, 'max_temp': max_temp, 'min_temp': min_temp}
    df = pd.DataFrame(data=d)

    df.to_csv('/network/tmp1/barinpvi/data/canopex_csv/'+f_name+'.csv', index=False)
```

```python
for f in glob.glob('/network/tmp1/barinpvi/data/canopex/CANOPEX_ENVCAN_ASCII/*.dly'):
    lines = correct_spaces(f)
    f_name = f.split('/')[-1].split('.')[0]
    print(f_name)
    preprocess(lines, f_name)
```
