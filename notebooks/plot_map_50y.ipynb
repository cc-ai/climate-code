```python
import rasterio as rio
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
from matplotlib.colors import LogNorm

plt.ion()

sns.set(font_scale=1.5)

sns.set_style("whitegrid")

plt.rcParams['figure.figsize'] = (8,8)

with rio.open('/network/home/barinpvi/floodMap/floodMapGL_rp50y.tif') as src:
    lidar_dem_im = src.read(1)
    lidar_dem_im[lidar_dem_im < 0] = np.nan
```

```python
lidar_dem_im.shape
```

```python
plt.imshow(lidar_dem_im)
plt.colorbar()
```

```python
plt.imshow(lidar_dem_im, norm=LogNorm())
plt.colorbar()
```

```python
plt.imshow(lidar_dem_im, norm=LogNorm(vmin=0.1, vmax=30))
plt.colorbar()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(lidar_dem_im, norm=LogNorm(vmin=0.1, vmax=30))
plt.colorbar()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(lidar_dem_im, norm=LogNorm(vmin=0.1, vmax=30))
plt.colorbar(orientation="horizontal",fraction=0.07,anchor=(1.0,0.0))
```
